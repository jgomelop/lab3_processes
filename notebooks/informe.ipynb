````python
import os
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Define the CSV file path
csv_file_path = os.path.join('../cpp_outputs/results.csv')
csv_file_path
````

````python
df = pd.read_csv(csv_file_path)

# Display the first few rows of the dataframe
print(df.head())
````

# Speedup vs Processes


````python
# Plot Speedup vs Processes
plt.plot(df['Processes'], df['Speedup'], marker='o', label='Speedup')
plt.xlabel('Processes')
plt.ylabel('Speedup')
plt.title('Speedup vs Processes')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig('speedup_vs_processes.png', dpi=300)  # Save the figure with high quality
plt.show()
````

# ParallelTime vs Processes

````python
# Plot ParallelTime vs Processes
plt.plot(df['Processes'], df['ParallelTime(s)'], marker='o', label='Parallel Time')
plt.xlabel('Processes')
plt.ylabel('Parallel Time (s)')
plt.title('Parallel Time vs Processes')
plt.legend()
plt.grid(True)
plt.savefig('parallel_time_vs_processes.png', dpi=300)  # Save the figure with high quality
plt.show()
````

# Time Comparison


````python
# Plot ParallelTime vs Processes with SequentialTime
plt.plot(df['Processes'], df['ParallelTime(s)'], marker='o', label='Parallel Time')
plt.plot(df['Processes'], df['SequentialTime(s)'], marker='o', label='Sequential Time')
plt.xlabel('Processes')
plt.ylabel('Total Time (s)')
plt.title('Execution Time Comparison')
plt.legend(loc='best')
plt.grid(True)
plt.savefig('execution_time_comparison.png', dpi=300)  # Save the figure with high quality
plt.show()
````
